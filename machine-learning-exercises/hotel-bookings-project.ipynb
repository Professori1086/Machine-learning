```python
# Importing the necessary libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px

# ML tools
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
hotel_bookings = pd.read_csv("hotel_bookings_clean.csv")
hotel_bookings.head()
```

## Ploting a bar graph showing the bookings per month

```python
bookings_by_month = hotel_bookings.groupby("arrival_date_month", as_index=False)[["hotel_Resort"]].count().rename(columns={"hotel_Resort":"nb_bookings"})
months = ["January","February","March","April","May","June","July","August","September","October",
          "November","December"]
fig = px.bar(bookings_by_month, x="arrival_date_month",y="nb_bookings",title="Hotel bookings by month",category_orders={"arrival_date_month":months})
fig.show(config={"displayModeBar":False})
```

```python
hotel_bookings.isna().sum()
```

```python
hotel_bookings.children = hotel_bookings.children.fillna(0)
```

```python
from sklearn.model_selection import KFold
# Splitting data
split = KFold(n_splits=4, shuffle=True, random_state=42)

# Creating a dictionary for the models
models = {"randomforest":RandomForestClassifier(n_jobs=-1, random_state=42),
         "decisiontree":DecisionTreeClassifier(random_state=42)}

#Creating x and y
X = hotel_bookings.drop("is_canceled", axis=1)
y = hotel_bookings["is_canceled"]

results = {}
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

for model_name, model in models.items():
    model.fit(X_train, y_train)
    results[model]= model.score(X_test,y_test)

results
```

```python
from sklearn.model_selection import cross_val_score
cv_results = cross_val_score(RandomForestClassifier(n_jobs=-1, random_state=42),X,y,scoring="precision")
cv_results
```

```python
precision = np.mean(cv_results)
precision
```

```python
r_score = np.mean(cross_val_score(RandomForestClassifier(n_jobs=-1, random_state=42),X,y,scoring="recall"))
r_score
```

```python
f1_score = np.mean(cross_val_score(RandomForestClassifier(n_jobs=-1, random_state=42),X,y,scoring="f1"))
```
